```python
def per_bor(precinct):
    if precinct <= 39:
        return "Manhattan"
    elif precinct >= 40 and precinct < 59:
        return "Bronx"
    elif precinct >=60 and precinct < 99:
        return "Brooklyn"
    elif precinct >= 100 and precinct < 119:
        return "Queens"
    elif precinct >= 120 and precinct < 125:
        return "Staten Island"
    else:
        return "unknown"
```

```python
police['borough'] = np.vectorize(per_bor)(police['precinct'])
```

```python
def mayor_yr(year):
    if year >= 1994 and year <= 2001:
        return "Giuliani"
    elif year >= 2002 and year <= 2013:
        return "Bloomberg"
    elif year >= 2014 and year <= 2020:
        return "DeBlasio"
    else:
        return "unknown"
```

```python
police['mayor'] = np.vectorize(mayor_yr)(police['year_received'])
police.head(5)
police['mayor'].value_counts(dropna=False)
```

```python
police = police[police.mayor != 'unknown']
police['mayor'].value_counts(dropna=False)
```

```python
police.loc[police['complainant_age_incident']== police['complainant_age_incident'].min()]
```

```python
police = police[(police['complainant_age_incident'] >0) | (police['complainant_age_incident'].isna())]
police.info()
```

```python
police = police[police['precinct'].notna()]
```

```python
police = police[(police['complainant_age_incident'] > 10) | (police['complainant_age_incident'].isna())]
# police.info()
```

```python
avg_age = police.complainant_age_incident.mean()
avg_age = avg_age.round(0)
avg_age
```

```python
police['complainant_age_incident'].fillna((police['complainant_age_incident'].mean()),inplace=True)
# sub2['income'].fillna((sub2['income'].mean()), inplace=True)
police.loc[police['complainant_age_incident'].isna()]
```
